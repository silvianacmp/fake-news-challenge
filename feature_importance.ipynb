```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')

agree = X_df[y==0]['cosine_tf_idf']
disagree = X_df[y==2]['cosine_tf_idf']
discuss = X_df[y==1]['cosine_tf_idf']
unrelated = X_df[y==3]['cosine_tf_idf']

weights = np.ones_like(agree)/float(len(agree))
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0, 0].hist(agree, bins=20, label='Agree', color='darkseagreen', weights=weights)
ax[0, 0].set_title('Agree cosine tf-idf')
ax[0, 0].set_xlim(0.0, 1.0)

weights = np.ones_like(disagree)/float(len(disagree))
ax[0, 1].hist(disagree, bins=20, label='Disagree', color='darksalmon', weights=weights)
ax[0, 1].set_title('Disagree cosine tf-idf')
ax[0, 1].set_xlim(0.0, 1.0)

weights = np.ones_like(discuss)/float(len(discuss))
ax[1, 0].hist(discuss, bins=20, label='Discuss', color='steelblue', weights=weights)
ax[1, 0].set_title('Discuss cosine tf-idf')
ax[1, 0].set_xlim(0.0, 1.0)

weights = np.ones_like(unrelated)/float(len(unrelated))
ax[1, 1].hist(unrelated, bins=20, label='Unrelated', color='dimgrey', weights=weights)
ax[1, 1].set_title('Unrelated cosine tf-idf')
ax[1, 1].set_xlim(0.0, 1.0)
plt.show()
```
